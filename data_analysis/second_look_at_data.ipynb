<h1>A Look Into COVID-19 Cases Between Ivy and Corresponding State Schools</h1>

```python
import pandas as pd
colleges_df = pd.read_csv('../data/colleges.csv')
```

```python
ivy_schools = ['Brown University', 'Columbia University', 'Cornell University', 'Dartmouth College', 'Harvard University', 'University of Pennsylvania', 'Princeton University', 'Yale University']
ivy_filter = colleges_df['college'].isin(ivy_schools) 
ivy_df = colleges_df[ivy_filter]
print(ivy_df)
```

This is the data of just ivy league schools pulled out from the NYT dataset. From this data, we can see that the ivy with the most COVID-19 cases is UPenn. Not only does UPenn have the most cases, in comparison to the other ivy leagues that come in around 100 and 200, UPenn has an astounding 660 cases. For the sake of counting colleges by state, I am combining Cornell and Columbia, since they are both in New York. Their combined number of cases make 283, which makes New York ivies the second highest amount of cases.

```python
SUNY = ['SUNY Albany',
        'SUNY Alfred State College of Technology',
        'SUNY Binghamton',
        'SUNY Brockport',
                  'SUNY Broome Community College',
                  'SUNY Buffalo State',
                  'SUNY Canton',
                  'SUNY Cobleskill',
                  'SUNY College at Potsdam',
                  'SUNY College of Environmental Science and Forestry',
                  'SUNY College of Optometry',
                  'SUNY Corning Community College',
                  'SUNY Cortland',
                  'SUNY Delhi',
                  'SUNY Downstate Health Sciences University',
                  'SUNY Empire State College',
                  'SUNY Geneseo',
                  'SUNY Maritime College',
                  'SUNY Morrisville',
                  'SUNY New Paltz',
                  'SUNY Old Westbury',
                  'SUNY Oneonta',
                  'SUNY Oswego',
                  'SUNY Plattsburgh',
                  'SUNY Polytechnic Institute',
                  'SUNY Purchase College',
                  'SUNY University at Buffalo',
                  'SUNY Westchester Community College',
                  'SUNY at Fredonia',]



umass = [          'University of Massachusetts Amherst', 
                  'University of Massachusetts Boston',
                  'University of Massachusetts Dartmouth'
                  'University of Massachusetts Lowell',
                  'University of Massachusetts Medical School',]

pennstate = ['Penn State University', 
                  'Penn State University Abington',
                  'Penn State University Altoona',
                  'Penn State University Beaver',
                  'Penn State University Behrend',
                  'Penn State University Berks',
                  'Penn State University Brandywine',
                  'Penn State University College of Medicine',
                  'Penn State University Dickinson Law',
                  'Penn State University DuBois',
                  'Penn State University Fayette',
                  'Penn State University Great Valley',
                  'Penn State University Greater Allegheny',
                  'Penn State University Harrisburg',
                  'Penn State University Hazleton',
                  'Penn State University Lehigh Valley',
                  'Penn State University Mont Alto',
                  'Penn State University New Kensington',
                  'Penn State University Schuylkill',
                  'Penn State University Scranton',
                  'Penn State University Shenango',
                  'Penn State University Wilkes-Barre',
                  'Penn State University York',]

uconn = ['Central Connecticut State University',
                  'Eastern Connecticut State University',
                  'Southern Connecticut State University',
                  'Western Connecticut State University']

state_colleges = ['University of Rhode Island', 
                  'Rutgers University', 'University of New Hampshire'] + SUNY + umass + uconn + pennstate

state__college_filter = colleges_df['college'].isin(state_colleges) 
state_colleges_df = colleges_df[state__college_filter]

```

```python
college2group = { n: n.split(' ')[0] for n in SUNY }

college2group.update({ n : 'UCONN' for n in uconn})
college2group.update({ n : 'UMASS' for n in umass})
college2group.update({ n : 'PENNSTATE' for n in pennstate})

college2group['University of Rhode Island']='University of Rhode Island'
college2group['Rutgers University']='Rutgers University'
college2group['University of New Hampshire']='University of New Hampshire'

college2group
```

```python
state_colleges_df['college_group']=state_colleges_df['college'].map(college2group).values
state_colleges_df
```

```python
state_colleges_df.groupby('college_group')['cases'].sum()
```

For the sake of comparing colleges at the state level, I am condensing different campuses of state schools into one state school entity. For example, the different Penn State campuses will count towards one Penn State group. The state schools seem to trend how the ivies are, with Penn State having the most cases and SUNY having the second highest. However, unlike with the ivy league, SUNY seen to have a comparable amount of cases to Penn State, even though both Cornell and Columbia combined still totaled less than half of UPenn's cases. 

```python
state_colleges_df.groupby('college_group')['cases'].mean()
```

So these numbers are actually a bit misleading because with Rutgers and the U of RI, we can't calculate means since they only have one campus, whereas the other state schools have multiple campuses. Excluding those schools, this data supports our previous data. We see Penn State having the most average amount of cases, as well as SUNY coming in second here as well.
