<h1>COVID-19 Cases in Ivy League Schools and Related Areas</h1>

Since March of 2020, the world has been devastated by the COVID-19 pandemic which has lead to countless losses. At first, most of the world did not take this virus seriously, since plagues have not had mass impact in major western countries for a long time. However, it soon became clear that COVID-19 was no trivial sickness, and that the way it traveled between people was quick and deadly: through water droplets. Even worse, the symptoms of infection could take up to two weeks to manifest, meaning that people who didn't think they were sick, could, in fact, pass on the virus to other people they meet 

As COVID-19 rampaged across the global, major health organizations encouraged people to practice social distancing, which meant staying at home unless going outside was absolutely necessary. Of course, people rebelled against this suggestion. The timing was just unfortunate; many students in universities and colleges were on their spring breaks. It's hard to keep to oneself when human beings are such social beings, especially halfway through a tough semester. Besides, at the start of the crisis, it was believed that only the elderly and the sick were in danger of dying from this virus. The President of the United States even encouraged the doubt of the severity of the situation, claiming that COVID-19 was similar to the yearly flu. Regrettably, that was proven not the case.

One of the things that stuck with me during this entire pandemic was of a college student saying that they'd party regardless of being infected with COVID-19 or not, since it was spring break. Soon after, reports upon reports demonstrated that this virus was not only lethal to the elderly and sickly, but could spell death for everyone it infected. Many schools shifted the rest of their spring semester online, wanting to minimize the contact students, faculty, and staff had with each other. Social distancing was preached, over and over again.

The spring semester and a summer went by. Promises about the pandemic were made and broken. Relief kept being pushed back as case numbers and deaths spiked. The estimated time to contain this virus was extended again and again. There was no end in sight, yet, the world went on. Summer would end and school would begin again.

Fingers were quickly pointed at college students returning to campus, since social distancing was near impossible with the amount of students universities and colleges boasted.  College students were accused of contributing to growing infection rates with stereotypical reasoning: partying and gathering in large numbers for no reason other than to simply enjoy life. As news coverage continue to single out college students over and over again, I began to wonder two things: If college students really were impacting case numbers, and if the class of college mattered at all to the contribution.    

Firstly, I want to define what I mean by "class" because I do not wish to offend anyone. Enrolling in any institute of higher education is a feat in itself. However, as a student in an ivy league school, I imagine that schools classified as "elite" schools would have more resources to mitigate the impact of a pandemic because they tend to generate more revenue. With this distinction, I began to compare ivy league schools to their corresponding state school. I also tried to compare the number of cases to the population of the school to see if the percentages revealed anything interesting.

<h2>Conclusion</h2>

Though COVID-19 is a deadly virus and all of us should be cautious in the mist of this pandemic, it is wrongful to paint college students as whole as people who are purposely gathering and spreading this disease. There is perhaps some truth to the carelessness of college students at the beginning of this pandemic, but since then, the number of cases of COVID-19 of college students cannot even compare to the number of state cases at large. As for the issue of handling COVID-19 between ivy league schools and state schools, I believe that the reason why state schools contribute a larger amount of cases to the total number of higher education COVID-19 cases is because state schools tend to be more widespread. In comparing the schools through number of cases and student population,  there is not really a trend that paints either category of schools as doing better, as some state schools have a higher percentage of cases per student population, and some ivy leagues as well.
